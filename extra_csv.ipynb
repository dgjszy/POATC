```python
sheet_names = xls.sheet_names
df = pd.read_excel(xls,sheet_name=sheet_names[-1])
```

```python
date = df.iloc[0,3].split('：')[-1]
date = date.replace('年', '.')
date = date.replace('月', '.')
```
